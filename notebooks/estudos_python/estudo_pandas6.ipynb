**PIVOT Tables com Pandas**

Uma PIVOT table serve para sumarizar características de um dataset maior

```python
# Para exemplificar vamos utilizar o dataset das primarias americanas de 2016
import pandas as pd
import numpy as np

p_results_df = pd.read_csv('res/primary-results.csv')
```

```python
# sumarizando informacoes sobre a soma dos votos por estado(state) partido(party) e candidatos(candidate)

pd.pivot_table(p_results_df,index=['state','party','candidate'],values=['votes'],aggfunc={'votes':np.sum}).head()
```

```python
# Para deixar a sumarização mais interessante vamos inserir uma coluna de rank (por distrito e partido) no dataframe original
p_results_df['rank']=p_results_df.groupby(['county','party'])['votes'].rank(ascending=False)

#Verificando os dados para o distrito de Los Angeles
p_results_df[p_results_df['county'] == 'Los Angeles']
```

```python
# sumarizando a o dataframe levando em conta a informacao do rank
pivot=pd.pivot_table(p_results_df,index=['state','county','party','candidate'],values=['votes','rank'],aggfunc={'votes':np.sum,'rank':np.min})

#contando o numero de vitorias que os candidatos de cada partido tiveram em cada distrito 
pivot[pivot['rank']==1.0].groupby(['candidate','party']).aggregate({'rank':np.sum})
```
